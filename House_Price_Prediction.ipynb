```python
housing_data = housing_data.drop(['statezip', 'country','street'], axis=1)
```

```python
housing_data = housing_data.drop(['city'], axis=1)
```

```python
housing_data['date'] = pd.to_datetime(housing_data['date'],format='mixed')
housing_data['year_present'] = housing_data['date'].dt.year
housing_data['building_age'] = housing_data['year_present'] - housing_data['yr_built']
```

```python
housing_data = housing_data.drop(['date','year_present','yr_built'], axis=1)
```

```python
housing_data.head()
```

```python
housing_data = housing_data.drop(['yr_renovated','waterfront','sqft_basement'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
# Assuming 'housing_data' is the name of your DataFrame
# Select the numerical columns to be normalized
numerical_columns = ['sqft_living', 'sqft_lot','sqft_above']  # Replace with actual numerical column names

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Normalize the selected columns
housing_data[numerical_columns] = scaler.fit_transform(housing_data[numerical_columns])
```

```python
corr_matrix=housing_data.corr(method="pearson")
plt.figure(figsize=(10,10))    
sns.heatmap(corr_matrix,annot=True,cmap='Blues')   
high_corr_feature=corr_matrix.index[abs(corr_matrix['price'])>=0.2].tolist()
high_corr_feature.remove('price')# Remove the target variable from the list   
# print(corr_matrix)
print(high_corr_feature) 
```

```python
X=housing_data[high_corr_feature]
Y=housing_data['price']
```

```python
from sklearn import linear_model
reg=linear_model.LinearRegression()
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.30)
```

```python
reg.fit(xtrain,ytrain)
```

```python
pred=reg.predict(xtest)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
r2score =  r2_score(ytest, pred)
print("r2_score:", r2score)
```

```python
mse=mean_squared_error(ytest,pred)
```
